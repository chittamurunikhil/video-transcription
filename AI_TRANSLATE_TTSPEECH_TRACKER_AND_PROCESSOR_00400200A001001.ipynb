```python
# Import libraries
import pytube

# Define function to download video

def download_video(url, output_path="."):
  """Downloads a YouTube video to the specified output path.

  Args:
      url: The URL of the YouTube video to download.
      output_path: The path to save the downloaded video (optional).
  """
  try:
    # Create a YouTube object
    youtube = pytube.YouTube(url)

    # Get the highest resolution stream (modify for specific resolution if needed)
    video_stream = youtube.streams.get_highest_resolution()

    # Download the video
    video_stream.download(output_path=output_path)

    print("Video downloaded successfully!")

    # Get video title
    video_title = youtube.title
    return video_title
  except Exception as e:
    print(f"An error occurred: {e}")
    return None

```

```python
# run only if text file saving is required
def save_text_to_txt(text, filename):
  """
  Saves text to a TXT file.

  Args:
      text: The text content to save.
      filename: The desired name (including path) for the TXT file.
  """
  with open(filename, 'w', encoding='utf-8') as f:
    f.write(text)

# Example usage
text =  text # This is the text content to be saved.
filename = "my_text.txt"
save_text_to_txt(text, filename)
print(f"Text saved to: {filename}")
```
